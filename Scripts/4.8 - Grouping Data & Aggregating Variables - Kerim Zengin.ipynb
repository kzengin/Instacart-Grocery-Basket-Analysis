# 2. In this Exercise, you learned how to find the aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe. Now, repeat this process for the entire dataframe. <a class="anchor" id="step2"></a>

# 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions. <a class="anchor" id="step4"></a>

```python
#Creating a max_order column to hold the maximum order numbers of each user, grouped by user_id
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Checking if previous code was executed successfully
ords_prods_merge.head()
```

```python
#Creating a flag for Loyal Customer for values higher than 40
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Creating a flag for Regular customer for values more than 10 and less than and equals 40
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#Creating a flag for New customer for values equal or less than 10
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Checking if loyalty flags were executed successfully
ords_prods_merge.head()
```

```python
#Checking the frequency counts for the loyalty_flag column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# 5. The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers. <a class="anchor" id="step5"></a>

```python
#Aggregating the prices' mean, min, max values and grouping results by loyalty_flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

#### This shows that on average new customers spend more than the loyal and regular customers

# 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
## If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.” <a class="anchor" id="step6"></a>

```python
#Creating an avg_price_orders column to hold the average prices of all orders for each user, grouped by user_id
ords_prods_merge['avg_price_orders'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Checking if previous code was executed successfully
ords_prods_merge.head()
```

```python
#Creating a flag for high spender for values higher than or equals to 10 and insert flag into spending_habit column
ords_prods_merge.loc[ords_prods_merge['avg_price_orders'] >= 10, 'spending_habit'] = 'High spender'
```

```python
#Creating a flag for low spender for values lower than 10 and insert flag into spending_habit column
ords_prods_merge.loc[ords_prods_merge['avg_price_orders'] < 10, 'spending_habit'] = 'Low spender'
```

```python
#Checking if spending_habit flags were executed successfully
ords_prods_merge.head()
```

```python
#Showing the frequency count for the spending_habit column
ords_prods_merge['spending_habit'].value_counts(dropna=False)
```

# 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.” <a class="anchor" id="step7"></a>

```python
#Creating an median_days_prior_order column to hold the median days since last order for each user, grouped by user_id
ords_prods_merge['median_days_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Checking if previous code was executed successfully
ords_prods_merge.head()
```

```python
#Creating a flag for frequent customers for median values lower than or equals to 10 and insert flag into order_frequency column
ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] <= 10, 'order_frequency'] = 'Frequent spender'
```

```python
#Creating a flag for frequent customers for median values higher than 10 and lower than and equals to 20 and insert flag into order_frequency column
ords_prods_merge.loc[(ords_prods_merge['median_days_prior_order'] > 10) & (ords_prods_merge['median_days_prior_order'] <= 20), 'order_frequency'] = 'Regular spender'
```

```python
#Creating a flag for frequent customers for median values higher than 20 and insert flag into order_frequency column
ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] > 20, 'order_frequency'] = 'Non-frequent spender'
```

```python
#Checking if previous code was executed successfully
ords_prods_merge.head()
```

```python
#Showing the frequency count for the order_frequency column
ords_prods_merge['order_frequency'].value_counts(dropna=False)
```
