```python
import pandas as pd
import numpy as np
import os
```

```python
#Creating a folderpath shortcut for importing/export dataframes
path = r'/Users/kerimzengin/Desktop/Data Analytics/Data Analytics Immersion/Achievement 4 - Python Fundamentals for Data Analysts/Instacart Basket Analysis/Data'
```

```python
#importing Orders/Products Merged pickle file into ords_prods_merged dataframe
ords_prods_merged = pd.read_pickle(os.path.join(path, 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
#create a subset of a million rows
df = ords_prods_merged[:1000000]
```

```python
#Showing the number of rows and columns for the df dataframe
df.shape
```

```python
#define a function for creating a pricing labeled column
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function to the df dataframe
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
#Showing the frequency counts for the 'price_range' column
df['price_range'].value_counts(dropna = False)
```

```python
#Showing the maximum value within the 'price' column
df['prices'].max()
```

```python
#Creating a loc function for the pricing labeled column for High-range products
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
##Creating a loc function for the pricing labeled column for Mid-range products
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
##Creating a loc function for the pricing labeled column for Low-range products
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Showing the frequency counts for the 'price_range_loc' column
df['price_range_loc'].value_counts(dropna = False)
```

```python
#Adding the price_range_loc function to the ords_prods_merged dataframe for the High-Range Products
ords_prods_merged.loc[ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-Range Product'
```

```python
#Adding the price_range_loc function to the ords_prods_merged dataframe for the Mid-Range Products
ords_prods_merged.loc[(ords_prods_merged['prices'] <= 15) & (ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-Range Product'
```

```python
#Adding the price_range_loc function to the ords_prods_merged dataframe for the Low-Range Products
ords_prods_merged.loc[ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-Range Product'
```

```python
#Showing the frequency counts for the newly created 'price_range_loc' column
ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

```python
#Showing the frequency counts for the 'order_days_of_week' column
ords_prods_merged['order_days_of_week'].value_counts(dropna = False)
```

```python
#Creating a list for the for-loop function 
result = []

for value in ords_prods_merged["order_days_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Showing the results for the result list following the for-loop
result
```

```python
#Creating a 'busiest_day' column with the result from the for-loop for the ords_prods_merged dataframe
ords_prods_merged['busiest_day'] = result
```

```python
#Showing the frequency counts for the 'busiest_day' column
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method. <a class="anchor" id="step2"></a>

```python
#Creating a function to define the 2 busiest and slowest days instead just one
result = []

for value in ords_prods_merged["order_days_of_week"]:
  if value == 0 and 1:
    result.append("Busiest day")
  elif value == 3 and 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#Creating a 'busiest_days' column with the result from the for-loop for the ords_prods_merged dataframe
ords_prods_merged['busiest_days'] = result
```

```python
#Creating a 'slowest_days' column with the result from the for-loop for the ords_prods_merged dataframe
ords_prods_merged['slowest_days'] = result
```

# 3. Check the values of this new column for accuracy. Note any observations in markdown format. <a class="anchor" id="step3"></a>

```python
#Showing the frequency counts for the 'busiest_days' & 'slowest_days' columns
ords_prods_merged['busiest_days' and 'slowest_days'].value_counts(dropna = False)
```

# 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.” <a class="anchor" id="step4"></a>

```python
#Showing the frequency counts for the 'order_hour_of_week' column to get the busiest hours
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
#Creating a function to define the busiest hours and slowest hours
orders = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value == 10 and 11 and 14 and 15 and 13 and 12 and 16 and 9:
    orders.append("Most Orders")
  elif value == 0 and 1 and 5 and 2 and 4 and 3:
    orders.append("Fewest Orders")
  else:
    orders.append("Average Orders")
```

```python
#Showing the results for the orders list
orders
```

```python
#Creating a 'busiest_period_of_day' column with the results from the for-loop for the ords_prods_merged dataframe
ords_prods_merged['busiest_period_of_day'] = orders
```

```python
#Checking if the busiest_period_of_day column was created successfully
ords_prods_merged.head()
```
