# 5. Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task. <a class="anchor" id="step5"></a>

```python
#Importing products_checked.csv into a dataframe
df_prods = pd.read_csv(os.path.join(path, 'Prepared Data', 'products_checked.csv'))
```

```python
#Checking the default number of records for the Products dataframe
df_prods.head()
```

```python
#Checking the number of rows and columns for the Products dataframe
df_prods.shape
```

```python
#Dropping the Unnamed: 0 column as it is not necessary
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
#Checking if the unnecessary column is removed
df_prods.head()
```

```python
#Checking the default records for the Combined Orders/Products dataframe
df_ords_prods_combined.head()
```

```python
#Dropping the _merge column from previous merging process
df_ords_prods_combined = df_ords_prods_combined.drop(columns = '_merge')
```

```python
#Dropping the Unnamed: 0 column as it is not necessary
df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['Unnamed: 0'])
```

```python
#Checking if _merge & Unnamed columns has dropped
df_ords_prods_combined
```

```python
#Dropping the Unnamed: 0.1 column as it is not necessary
df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['Unnamed: 0.1'])
```

```python
#Checking if _merge & Unnamed columns has dropped
df_ords_prods_combined
```

```python
#Merging the Products dataframe onto the Orders/Products combined dataframe into a new merged dataframe
df_ords_prods_merged = df_ords_prods_combined.merge(df_prods, on = 'product_id', how = 'left', indicator = True)
```

```python
#Checking if the merge was successful
df_ords_prods_merged.head()
```

```python
#Checking the number of rows and columns after merging
df_ords_prods_merged.shape
```
