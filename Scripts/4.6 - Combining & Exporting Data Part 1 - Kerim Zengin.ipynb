# Importing Libraries and Dataframes <a class="anchor" id="import_libraries"></a>

```python
import pandas as pd
import numpy as np
import os
```

```python
#Creating path variable for shortcuting folderpath for importing/exporting
path = r'/Users/kerimzengin/Desktop/Data Analytics/Data Analytics Immersion/Achievement 4 - Python Fundamentals for Data Analysts/Instacart Basket Analysis/Data'
```

```python
#Importing orders_checked.csv as df_ords dataframe
df_ords = pd.read_csv(os.path.join(path, 'Prepared Data', 'orders_checked1.csv'), index_col = False)
```

```python
#Importing order_products__prior.csv as df_ords_prior dataframe
df_ords_prior = pd.read_csv(os.path.join(path, 'Original Data', 'order_products__prior.csv'), index_col = False)
```

# Merging Orders Prior dataframe with Orders dataframe <a class="anchor" id="merge"></a>

```python
#Check the output for df_ords_prior dataframe
df_ords_prior.head()
```

```python
#Check the output for df_ords dataframe
df_ords.head(30)
```

```python
#Checking the shape of Orders dataframe
df_ords.shape
```

```python
#Merging Orders Prior dataframe with Orders dataframe
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
#Checking if the merge was completed successfully
df_merged_large.head()
```

```python
#Checking the frequency count for the _merge column if full merge
df_merged_large['_merge'].value_counts()
```
